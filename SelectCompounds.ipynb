# Выбираем те что с F, но без Fe

При этом правда отсеется что-то что содержит и F и Fe, но таких явно о-малое

```python
compounds = pd.read_csv("csvs/compound_structures.csv")
compounds
```

```python
with_f = compounds[compounds.canonical_smiles.str.contains('F')]
with_f = with_f[~with_f.canonical_smiles.str.contains('Fe')]
with_f = with_f.drop(columns=["molfile", "standard_inchi", "standard_inchi_key", ])
with_f
```

```python
def siles_F_to_H(x) -> str:
    a = str.replace(x, 'F', '')
    return str.replace(a, '()', '')
def smiles_count_F(x) -> int:
    return x.count('F')
with_f['Smiles_H'] = with_f['canonical_smiles'].apply(siles_F_to_H)
with_f['count_F'] = with_f['canonical_smiles'].apply(smiles_count_F)
with_f
```
