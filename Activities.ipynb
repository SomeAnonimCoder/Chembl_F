# Загрузка данных

Итак, у нас есть пары molregno : molregno со фтором

```python
with_f = pd.read_csv("F_Hpairs.csv")[['molregno', 'molregno_H']]
with_f
```

# Читаем активности, пусть будут IC50

Читаем из файла, выкидываем лишнее, остается id, id работы, molregno и значение.

```python
activities = pd.read_csv('csvs/activities.csv')
activities = activities[activities['standard_type']=='IC50']
activities = activities[['activity_id','assay_id', 'molregno', 'value']]
activities
```

```python
with_f_act = with_f.merge(activities, on=["molregno"])
with_f_act
```

```python
with_f_act['molregno_F'] = with_f_act.molregno
with_f_act['molregno'] = with_f_act.molregno_H
with_f_act['assay_F'] = with_f_act.assay_id
with_f_act['value_F'] = with_f_act.value
with_f_act = with_f_act[['molregno', 'molregno_F', 'assay_F', 'value_F']]
with_f_act
```

```python
with_f_act = with_f_act.merge(activities, on=["molregno"])
with_f_act['molregno_H'] = with_f_act.molregno
with_f_act['assay_H'] = with_f_act.assay_id
with_f_act['value_H'] = with_f_act.value
with_f_act = with_f_act[['molregno_F', 'assay_F', 'value_F','molregno_H', 'assay_H', 'value_H']]
with_f_act
```

```python
activities_clean = with_f_act[with_f_act.assay_F==with_f_act.assay_H]
activities_clean
```

```python
activities_clean.to_csv("activities_clean.csv")
```
